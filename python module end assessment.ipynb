```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data frame from the source path
df = pd.read_excel("C:/Users/Lenovo/Downloads/myexcel.xlsx")
```

```python
df
```

```python
#assigning new random values for 'height' column
df['Height'] = np.random.randint(150, 180, size=len(df))
```

```python
df
```

```python
#checking for null values
df.isnull().sum()
```

```python
#checking for duplicate values
df.duplicated().sum()
```

```python
#assigning the value as 'unknown' for null values in the column 'college'
df['College'] = df['College'].fillna('Unknown')
```

```python
#assigning mean value to the null columns in the 'salary' column

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())



```

```python
df
```

```python
##Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.
team_counts = df['Team'].value_counts()
total_employees = len(df)
team_percentage = (team_counts / total_employees) * 100

team_distribution = pd.DataFrame({
    'Number of Employees': team_counts,
    'Percentage Split (%)': team_percentage
})
    
print(team_distribution)                    
```

```python
position_counts = df['Position'].value_counts()
position_percentages = (position_counts / total_employees) * 100
```

```python
employees_by_position = df.groupby('Position')['Name'].apply(list).reset_index()
print(employees_by_position)
```

```python
bins = [20, 30, 40, 50, 60]
labels = ['20-30', '30-40', '40-50', '50-60']

# Add a column with the age group
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_counts = df['Age_Group'].value_counts()

print(age_group_counts)


# Find the predominant age group
predominant_age_group = age_group_counts.idxmax()
predominant_count = age_group_counts.max()

print(f"The predominant age group is: {predominant_age_group} with {predominant_count} employees.")
```

```python
team_salary = df.groupby('Team')['Salary'].sum()
position_salary = df.groupby('Position')['Salary'].sum()

highest_team = team_salary.idxmax()
highest_team_salary = team_salary.max()

highest_position = position_salary.idxmax()
highest_position_salary = position_salary.max()

print(f"Team with highest salary expenditure: {highest_team} (${highest_team_salary:,.2f})")
print(f"Position with highest salary expenditure: {highest_position} (${highest_position_salary:,.2f})")
```

```python
df = df[['Age', 'Salary']].dropna()
df = df.astype({'Age': 'float', 'Salary': 'float'})

# Compute correlation
correlation = df.corr().loc['Age', 'Salary']
print(f"Correlation between Age and Salary: {correlation:.2f}")

# Plot scatter plot with regression line
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Age'], y=df['Salary'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title(f'Age vs Salary (Correlation: {correlation:.2f})')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
#1   pie chart for how it distributed employees across each team

plt.figure(figsize=(8, 8))
team_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Pastel1')
plt.title('Employee Distribution Across Teams')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
#2   barchart for "Segregation of employees based on their positions"

plt.figure(figsize=(10, 6))
sns.barplot(x=position_counts.index, y=position_counts.values, hue=position_counts.index, dodge=False, palette='viridis', legend=False)
plt.title('Employee Distribution Across Positions')
plt.xlabel('Position')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#3  Plot age group distribution as a bar chart

plt.figure(figsize=(8, 6))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, hue=age_group_counts.index, palette='coolwarm', legend=False)
plt.title('Distribution of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#4  graphical representation to Discover which team and position have the highest salary expenditure.

plt.figure(figsize=(10, 6))
sns.barplot(x=team_salary.index, hue=team_salary.index,y=team_salary.values, palette='coolwarm',legend=False)
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#5
plt.figure(figsize=(8, 5))
sns.regplot(x='Age', y='Salary', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title(f'Age vs Salary (Correlation: {correlation:.2f})')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
